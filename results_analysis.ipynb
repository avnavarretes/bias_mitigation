```python
linear_gender = linear[['prompt','gender']].groupby('prompt').value_counts(normalize=True)
poly_gender = poly[['prompt','gender']].groupby('prompt').value_counts(normalize=True)
constant_gender = constant[['prompt','gender']].groupby('prompt').value_counts(normalize=True)

linear_gender2 = linear2[['prompt','gender']].groupby('prompt').value_counts(normalize=True)
poly_gender2 = poly2[['prompt','gender']].groupby('prompt').value_counts(normalize=True)
constant_gender2 = constant2[['prompt','gender']].groupby('prompt').value_counts(normalize=True)
```

```python
linear_race = linear[['prompt','race']].groupby('prompt').value_counts(normalize=True)
poly_race = poly[['prompt','race']].groupby('prompt').value_counts(normalize=True)
constant_race = constant[['prompt','race']].groupby('prompt').value_counts(normalize=True)

linear_race2 = linear2[['prompt','race']].groupby('prompt').value_counts(normalize=True)
poly_race2 = poly2[['prompt','race']].groupby('prompt').value_counts(normalize=True)
constant_race2 = constant2[['prompt','race']].groupby('prompt').value_counts(normalize=True)
```

```python
constant_race2
```

```python
linear_gender = pd.DataFrame(linear_gender).reset_index()
linear_gender.columns = ['prompt','gender','linear300']

linear_gender2 = pd.DataFrame(linear_gender2).reset_index()
linear_gender2.columns = ['prompt','gender','linear20']

poly_gender = pd.DataFrame(poly_gender).reset_index()
poly_gender.columns = ['prompt','gender','poly300']

poly_gender2 = pd.DataFrame(poly_gender2).reset_index()
poly_gender2.columns = ['prompt','gender','poly20']

constant_gender = pd.DataFrame(constant_gender).reset_index()
constant_gender.columns = ['prompt','gender','constant300']

constant_gender2 = pd.DataFrame(constant_gender2).reset_index()
constant_gender2.columns = ['prompt','gender','constant20']
```

```python
linear_race = pd.DataFrame(linear_race).reset_index()
linear_race.columns = ['prompt','race','linear300']

linear_race2 = pd.DataFrame(linear_race2).reset_index()
linear_race2.columns = ['prompt','race','linear20']

poly_race = pd.DataFrame(poly_race).reset_index()
poly_race.columns = ['prompt','race','poly300']

poly_race2 = pd.DataFrame(poly_race2).reset_index()
poly_race2.columns = ['prompt','race','poly20']

constant_race = pd.DataFrame(constant_race).reset_index()
constant_race.columns = ['prompt','race','constant300']

constant_race2 = pd.DataFrame(constant_race2).reset_index()
constant_race2.columns = ['prompt','race','constant20']
```

```python
sdm_gender = sdm[['prompt','gender']].groupby('prompt').value_counts(normalize=True)
sdm_gender = pd.DataFrame(sdm_gender).reset_index()
sdm_gender.columns = ['prompt','gender','SDM']
```

```python
sdm_race = sdm[['prompt','race']].groupby('prompt').value_counts(normalize=True)
sdm_race = pd.DataFrame(sdm_race).reset_index()
sdm_race.columns = ['prompt','race','SDM']
```

```python
linear_df = linear[['prompt','gender','race']].groupby('prompt').value_counts(normalize=True)
linear_df = pd.DataFrame(linear_df).reset_index()
linear_df.columns = ['prompt','gender','race','linear300']
```

```python
linear_df2 = linear2[['prompt','gender','race']].groupby('prompt').value_counts(normalize=True)
linear_df2 = pd.DataFrame(linear_df2).reset_index()
linear_df2.columns = ['prompt','gender','race','linear20']
```

```python
poly_df = poly[['prompt','gender','race']].groupby('prompt').value_counts(normalize=True)
poly_df = pd.DataFrame(poly_df).reset_index()
poly_df.columns = ['prompt','gender','race','poly300']
```

```python
poly_df2 = poly2[['prompt','gender','race']].groupby('prompt').value_counts(normalize=True)
poly_df2 = pd.DataFrame(poly_df2).reset_index()
poly_df2.columns = ['prompt','gender','race','poly20']
```

```python
constant_df = constant[['prompt','gender','race']].groupby('prompt').value_counts(normalize=True)
constant_df = pd.DataFrame(constant_df).reset_index()
constant_df.columns = ['prompt','gender','race','constant300']
```

```python
constant_df2 = constant2[['prompt','gender','race']].groupby('prompt').value_counts(normalize=True)
constant_df2 = pd.DataFrame(constant_df2).reset_index()
constant_df2.columns = ['prompt','gender','race','constant20']
```

```python
sdm_df = sdm[['prompt','gender','race']].groupby('prompt').value_counts(normalize=True)
sdm_df = pd.DataFrame(sdm_df).reset_index()
sdm_df.columns = ['prompt','gender','race','SDM']
```

```python
summary_df = sdm_df.merge(linear_df, on=['prompt','gender','race'], 
                  how='outer').merge(linear_df2, on=['prompt','gender','race'],
                    how='outer').merge(poly_df, on=['prompt','gender','race'], 
                        how='outer').merge(poly_df2, on=['prompt','gender','race'], 
                            how='outer').merge(constant_df, on=['prompt','gender','race'], 
                                how='outer').merge(constant_df2, on=['prompt','gender','race'], how='outer').fillna(0).sort_values(['race','gender'])
```

```python
summary_gender = sdm_gender.merge(linear_gender, on=['prompt','gender'], 
                  how='outer').merge(linear_gender2, on=['prompt','gender'],
                    how='outer').merge(poly_gender, on=['prompt','gender'], 
                        how='outer').merge(poly_gender2, on=['prompt','gender'], 
                            how='outer').merge(constant_gender, on=['prompt','gender'], 
                                how='outer').merge(constant_gender2, on=['prompt','gender'], how='outer').fillna(0).sort_values('gender')
```

```python
linear_gender
```

```python
summary_race = sdm_race.merge(linear_race, on=['prompt','race'], 
                  how='outer').merge(linear_race2, on=['prompt','race'],
                    how='outer').merge(poly_race, on=['prompt','race'], 
                        how='outer').merge(poly_race2, on=['prompt','race'], 
                            how='outer').merge(constant_race, on=['prompt','race'], 
                                how='outer').merge(constant_race2, on=['prompt','race'], how='outer').fillna(0).sort_values('race')
```

```python
fig = plt.figure(figsize=(15,10))
fig.subplots_adjust(hspace=0.6, wspace=0.4)
prompt_list = ['ceo', 'executive', 'plumbing', 'doctor', 'nice', 'nurse', 'criminal','delivery','entrepreneur']
for i in range(0,9):
    ax = fig.add_subplot(3, 3, i+1)
    sns.barplot(x="variable", y="value", hue='race', data= pd.melt(summary_race2[summary_race2.prompt==prompt_list[i]].drop(columns=['prompt']), id_vars=["race"]), palette='pastel')
    plt.xticks(rotation = 45)
    plt.title(f'{prompt_list[i]}')
    plt.xlabel('')
    plt.ylabel('Percentage')
    plt.legend([],[], frameon=False)
plt.show()
```
